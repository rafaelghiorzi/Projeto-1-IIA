```python
import pandas as pd
```

```python
# Extraindo os dados para tratamento
mercados = pd.read_csv("farmers_markets_from_usda.csv")
vegetais = pd.read_csv("vegetables.csv")
```

```python
# Formatando os dados e excluindo informações desnecessárias
mercados.drop(columns=['FMID', 'OtherMedia', 'Season1Date','zip', 'Season2Date', 'Season3Date', 'Season4Date', 'Season1Time', 'Season2Time', 'Season3Time', 'Season4Time', 'Credit', 'WIC', 'WICcash', 'SFMNP', 'SNAP'], inplace=True)
vegetais.drop(columns=['Vegetable ID', 'Scientific Name', 'Color', 'Nutritional Value (per 100g)', 'Shelf Life (days)', 'Storage Requirements', 'Growing Conditions', 'Health Benefits', 'Common Varieties'], inplace=True)
# Remove dolar sign and convert to float on price
vegetais['Price (per kg)'] = vegetais['Price (per kg)'].replace({'\$': '', ',': ''}, regex=True).astype(float)
mercados = mercados.replace({'Y': True, 'N': False})
```
